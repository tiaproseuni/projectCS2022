## 12. Find for duplicates in the dogs dataset. Two records are duplicates if they have (1) same breeds and sex, and (2) they share at least 90% of the words in the description field. Extra points if you find and implement a more refined for determining if two rows are duplicates.

```python
from difflib import SequenceMatcher
```

```python
dogs_id = dogsOK.reset_index()
```

```python
def str_toset(string):
    sett1=set(string.split(' '))
    s1=sorted(s.strip(' .,:-"') for s in sett1)
    return s1
```

```python
 def share_desc(stringa_1, stringa_2):
        st1=str_toset(stringa_1)
        st2=str_toset(stringa_2)
        val=0
        if len(st2)>len(st1):
            for e in st2:
                if e in st1:
                    val+=1
        else:
            for e in st1:
                if e in st2:
                    val+=1
        return val/min([len(st1), len(st2)])
```

```python
def share_desc2(str1, str2):
    return SequenceMatcher(None, str1, str2).ratio()
```

```python
#esempio di funzionamento della prima funzione

desc1=dogs_id.loc[dogs_id['id']==46040566, 'description'].values[0]
desc2=dogs_id.loc[dogs_id['id']==46023195, 'description'].values[0]
print(desc1, "\n","\n","\n", desc2)
share_desc2(desc1, desc2)
```

```python
#esempio di funzionamento della seconda funzione

s = SequenceMatcher(None, desc1, desc2)
print(desc1, "\n","\n","\n", desc2)
s.ratio()
```

```python
def same_dog(id1, id2):
    desc1=dogs_id.loc[dogs_id['id']==id1, 'description'].values[0]
    desc2=dogs_id.loc[dogs_id['id']==id2, 'description'].values[0]
    
    if (type(desc1)!=type('')) | (type(desc2)!=type('')):
        return False
    else:
        pt = share_desc2(desc1, desc2)
        if pt >= 0.9:
            return True
        else:
            return False
```

```python
s=set()
d=dict()
dupl=set() #non serve
d_dupl=dict()

for index, row in dogs_id.iterrows():
    p_brd, s_brd, sex = row['breed_primary'], row['breed_secondary'], row['sex']
    if ((p_brd, s_brd, sex) not in s) | ((s_brd, p_brd, sex) not in s) :
        s.add((p_brd, s_brd, sex))
        d[(p_brd, s_brd, sex)]=[row['id']]
    else:
        dupl.add((p_brd, s_brd, sex)) #non serve
        d[(p_brd, s_brd, sex)].append(row['id'])

for el in d:
     if len(d[el])>1:
        d_dupl[el]=d[el] 
#ho creato un dizionario con per ogni tripletta (prim. breed, sec. breed, sex) che abbia un duplicato all'interno del df,
#associa i cani che hanno queste due breed e stesso sex
```

```python
def find_dupl(tripla):
    i=1
    lun=len(d_dupl[tripla])
    #br_pr, br_sec, sex = tripla
    for id_1 in d_dupl[tripla][:lun-1]:
        for id_2 in d_dupl[tripla][i:]:
            if id_1!=id_2:  #non serve
                if same_dog(id_1, id_2):
                    l.append((id_1, id_2))
        i+=1
```
